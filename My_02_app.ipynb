#Welcome to a Spiderman classifier app!!
If you have a photo of a Spiderman, and you are not sure whose the actor under the mask, try uploading your image and let us make a guess!

```python
from fastai.vision.all import *
from fastai.vision.widgets import *
```

```python
path = Path()
learn_inf = load_learner(path/'export.pkl')
```

```python
#Define upload button
btn_upload = widgets.FileUpload()

#Define visualization window
out_pl = widgets.Output()
out_pl.clear_output()

#Define results label
lbl_pred = widgets.Label()

#Define classification button
btn_run = widgets.Button(description='Classify')
```

```python
def on_click_classify(change):
    img = PILImage.create(btn_upload.data[-1])
    out_pl.clear_output()
    with out_pl: display(img.to_thumb(128,128))
    pred,pred_idx,probs = learn_inf.predict(img)
    lbl_pred.value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'

btn_run.on_click(on_click_classify)
```

```python
VBox([widgets.Label('Select your Spiderman!'), 
      btn_upload, btn_run, out_pl, lbl_pred])
```
